### 퍼셉트론

```
def perceptron(x1, x2):
    
    #가중치 w1, w2, 편향 w0
    w1, w2, w0 = 0.5, 0.5, 0.7 #AND
    #w1, w2, w0 = 1.0, 1.0, 0.5 #OR
    #w1, w2, w0 = -1.0, -1.0, -0.5 #NAND

    s = x1*w1 + x2*w2

    if s <= w0:
        return 0
    elif s > w0:
        return 1
```

```
import numpy as np
import matplotlib.pylab as plt 

x = np.array([[0,0], [0,1],[1,0],[1,1]])

plt.ylim(-0.1, 1.1)


for i in x:
    y= perceptron(i[0], i[1])
    
    if y == 0:
        alp = 0.2
    else:
        alp = 1

    plt.scatter(i[0], i[1], s=100, c="blue", alpha=alp,linewidths=4,edgecolors="blue")
    print (i[0], i[1],'->' , y)
    
plt.show()
```

## 활성화 함수

```
#계단 함수
import numpy as np
import matplotlib.pylab as plt

def step_function(x):
  y = x > 0
  return y.astype(np.int)

x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```
#시그모이드 함수
import numpy as np

def sigmoid(x):
  return 1 / (1+np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```
#ReLU함수
import numpy as np
import matplotlib.pylab as plt

def relu(x):
  return np.maximum(0, x)

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.show()
```
